Creando primero los datos

```python
import numpy as np 
from random import randint
from sklearn.utils import shuffle
from sklearn.preprocessing import MinMaxScaler
```

```python
train_lbl = []
train_smpl = []
```

```python

for i in range(50):
    # 5% of young individuals who experienced a certain effect
    random_young = randint(13,64)
    train_smpl.append(random_young)
    train_lbl.append(1)
    # 5% of older individuals who didnt experienced a certain effect
    random_older = randint(65,100)
    train_smpl.append(random_older)
    train_lbl.append(0)
```

```python
for i in range(1000): 
    random_young = randint(13,64)
    train_smpl.append(random_young)
    train_lbl.append(0)

    random_older = randint(65,100)
    train_smpl.append(random_older)
    train_lbl.append(1)
```

```python
for i in train_smpl:
    print(i)
```

```python
for i in train_lbl:
    print(i)
```

```python
train_lbl = np.array(train_lbl)
train_smpl = np.array(train_smpl)
train_lbl, train_smpl = shuffle(train_lbl, train_smpl)
```

```python
scaler = MinMaxScaler(feature_range=(0,1))
scaled_train_samples = scaler.fit_transform(train_smpl.reshape(-1,1))
```

```python
for i in scaled_train_samples:
    print(i)
```

Implementando un modelo secuencial con tf.keras

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Activation, Dense
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.metrics import categorical_crossentropy
```

```python
#GPU config was skipped
l1 = Dense(units=16, input_shape=(1,), activation='relu')
l2 = Dense(units=32 , activation='relu')
l3 = Dense(units=2, activation='softmax')
model = Sequential([l1,l2,l3])
```

```python
model.summary()
```

```python
model.compile(optimizer=Adam(learning_rate=0.0001), loss = 'sparse_categorical_crossentropy', metrics = ['accuracy'])
```

```python
model.fit(x=scaled_train_samples, y=train_lbl, validation_split=0.1, batch_size=10, epochs=30, shuffle=True, verbose=2)
```

Generando datos de prueba

```python
test_lbl = []
test_smpl = []
```

```python
for i in range(10):
    # 5% of young individuals who experienced a certain effect
    random_young = randint(13,64)
    test_smpl.append(random_young)
    test_lbl.append(1)
    # 5% of older individuals who didnt experienced a certain effect
    random_older = randint(65,100)
    test_smpl.append(random_older)
    test_lbl.append(0)
```

```python
for i in range(200): 
    # 5% of young individuals who experienced a certain effect
    random_young = randint(13,64)
    test_smpl.append(random_young)
    test_lbl.append(0)
    # 5% of older individuals who didnt experienced a certain effect
    random_older = randint(65,100)
    test_smpl.append(random_older)
    test_lbl.append(1)
```

```python
test_lbl = np.array(test_lbl)
test_smpl = np.array(test_smpl)
test_lbl, test_smpl = shuffle(test_lbl, test_smpl)
```

```python
scaled_test_samples = scaler.fit_transform(test_smpl.reshape(-1,1))
```

Prediction

```python
pred = model.predict(x=scaled_test_samples, batch_size=10, verbose=2)
```

```python
for p in pred:
    print(p)
```

```python
rounded_pred = np.argmax(pred, axis=-1)
```

```python
for i in rounded_pred:
    print(i)
```

```python
%matplotlib inline
from  sklearn.metrics import confusion_matrix
import itertools
import matplotlib.pyplot as plt
```

```python
conf_mat = confusion_matrix(y_true=test_lbl, y_pred=rounded_pred)
```

```python
def plot_conf_mat(cm, classes, normalize=False, title='Matriz de confusin', cmap = plt.cm.Blues):
    plt.imshow(cm, interpolation='nearest', cmap = cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)

    if normalize:
        cm = cm.astype('float')/cm.sum(axis=1)[:,np.newaxis]
        print('Normalized confusion matrix')
    else: 
        print('Confusion matrix, without normalization')

    print(cm)
    thresh = cm.max()/2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j,i, cm[i,j], horizontalalignment='center', color='white' if cm [i,j]>thresh else 'black')
    #plt.tight_layout()
    plt.ylabel('True lbl')
    plt.xlabel('Pred lbl')
```

```python
cm_plot_lbl = ['no_side_effects', 'had_side_effects']
plot_conf_mat(conf_mat, cm_plot_lbl)
```
